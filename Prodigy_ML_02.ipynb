```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
path='/content/drive/MyDrive/Mall_Customers.csv'
df = pd.read_csv(path)
df.head()
```

```python
plt.scatter(df.Age,df['Annual Income (k$)'])
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
```

```python
km = KMeans(n_clusters= 7)
```

```python
y_predicted = km.fit_predict(df[['Age','Annual Income (k$)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.drop('cluster',axis='columns')
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
df6= df[df.cluster==5]
df7= df[df.cluster==6]
plt.scatter(df1.Age,df1['Annual Income (k$)'],color='green')
plt.scatter(df2.Age,df2['Annual Income (k$)'],color='red')
plt.scatter(df3.Age,df3['Annual Income (k$)'],color='black')
plt.scatter(df4.Age,df4['Annual Income (k$)'],color='yellow')
plt.scatter(df5.Age,df5['Annual Income (k$)'],color='purple')
plt.scatter(df6.Age,df6['Annual Income (k$)'],color='orange')
plt.scatter(df7.Age,df7['Annual Income (k$)'],color='blue')
```

**MIN AND MAX**

```python
scaler = MinMaxScaler()
scaler.fit(df[['Annual Income (k$)']])
df['Annual Income (k$)']=scaler.transform(df[['Annual Income (k$)']])
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df.head()
```

```python
plt.scatter(df.Age,df['Annual Income (k$)'])
```

```python
km = KMeans(n_clusters=7)
y_predicted = km.fit_predict(df[['Age','Annual Income (k$)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
df6= df[df.cluster==5]
df7= df[df.cluster==6]
plt.scatter (df1.Age,df1['Annual Income (k$)'],color='green')
plt.scatter (df2.Age,df2['Annual Income (k$)'],color='purple')
plt.scatter(df3.Age,df3['Annual Income (k$)'],color='black')
plt.scatter(df4.Age,df4['Annual Income (k$)'],color='blue')
plt.scatter(df5.Age,df5['Annual Income (k$)'],color='pink')
plt.scatter(df6.Age,df6['Annual Income (k$)'],color='yellow')
plt.scatter(df7.Age,df7['Annual Income (k$)'],color='orange')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='maroon',marker='*',label='centroid')

plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
```

**ELBOW PLOT**

```python
sse = []
k_rng = range(1,200)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Annual Income (k$)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
